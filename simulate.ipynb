Graph fraction of agents conforming to platform's opinion against level of platform bias

```python
import json

with open('data/platform-vs-polarisation/data-b0-p0.5-c0.1-rec1.json') as f:
	data0 = json.load(f)['data']

with open('data/platform-vs-polarisation/data-b0.5-p0.5-c0.1-rec1.json') as f:
	data1 = json.load(f)['data']

# data0['data']
```

```python
def get_data(data):
	xs = []
	ys = []
	for key, val in data.items():
		xs.append(float(key))
		y = []
		for v in val:
			y.append(float(v[1] if v[0] == 1 else v[2]))
		ys.append(sum(y) / len(y))
	return xs, ys

xs0, ys0 = get_data(data0)
xs1, ys1 = get_data(data1)
```

```python
import matplotlib.pyplot as plt

plt.plot(xs0, ys0)
plt.plot(xs1, ys1)
```
